## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по правилам:

- если источник traffic_source равен Yandex или Google, то в source_type ставится organic;
- для источников paid и email из России ставим ad;
- для источников paid и email не из России ставим other;
- все остальные варианты берём из traffic_source без изменений.

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep = ';')
log.head()
```

```python
def filter(value):
    if value['traffic_source'] == 'yandex' or value['traffic_source'] == 'google':
        return 'organic'
    elif value['traffic_source'] == 'paid' or value['traffic_source'] == 'email':
        if value['region'] == 'Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return value['traffic_source']
```

```python
log['source_type'] = log.apply(filter, axis=1)
log.head()
```

```python
log[log['source_type'] == 'ad'].head()
```

```python
log[log['source_type'] == 'other'].head()
```

## Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
count_ratings = ratings.groupby('userId').count().reset_index()
count_ratings = count_ratings[['userId','rating']]
count_ratings.rename(columns = {'rating': 'amount_of_ratings'}, inplace=True)
count_ratings.head()
```

```python
count_ratings_more_than_100 = count_ratings[count_ratings['amount_of_ratings'] > 100]
count_ratings_more_than_100.head()
```

```python
max_timestamp = ratings.groupby('userId').max().reset_index()
min_timestamp = ratings.groupby('userId').min().reset_index()
```

```python
max_timestamp = max_timestamp[['userId','timestamp']]
min_timestamp = min_timestamp[['userId','timestamp']]
```

```python
max_timestamp.rename(columns = {'timestamp': 'max_timestamp'}, inplace = True)
min_timestamp.rename(columns = {'timestamp': 'min_timestamp'}, inplace = True)
```

```python
max_timestamp.head()
```

```python
min_timestamp.head()
```

```python
avg_lifetime = count_ratings_more_than_100.merge(max_timestamp, how='inner', on = 'userId').merge(min_timestamp, how='inner', on = 'userId')
```

```python
def avg(line):
    avg_timestamp = line['max_timestamp']- line['min_timestamp']
    return avg_timestamp
```

```python
avg_lifetime['mean'] = avg_lifetime.apply(avg, axis = 1)
avg_lifetime.head()
```
