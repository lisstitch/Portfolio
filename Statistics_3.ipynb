Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.

999: 100 руб.

555: 50 руб.

333: 15 руб.

111: 10 руб.

*77: 5 руб.

**7: 3 руб.

*00: 2 руб.

**0: 1 руб.

Вопросы:

- Выгодна ли игра игроку?

- Сколько в среднем приобретает или теряет игрок за одну игру?

**Дополнительное задание повышенной сложности.**

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

- любая цифра

## Основная часть задания

```python
import numpy as np
```

```python
win = {777: 200, 999: 100, 555: 50, 333: 15, 111: 10}
```

```python
def get_number():
    return np.random.randint(0, 1000)

def tails(number):
    if number % 100 == 77:
        return 5
    elif number % 10 == 7:
        return 3
    elif number % 100 == 0:
        return 2
    elif number % 10 == 0:
        return 1                                           
    return 0

def main():
    trials = 1000000
    successes = 0
    won = 0
    lost = 0 

    for i in range(trials):
        number = get_number()
        if number in win:
            successes += 1
            won += win[number]
        elif number not in win:
            tail = tails(number)
            if tail == 0:
                successes += 0
                won += tail
            else:
                successes += 1
                won += tail
        lost += 1

    print('Выигранно: ', won, 'руб.')
    print('Проигранно: ', lost, 'руб.')
    print('Средний выигрыш: ', round(won/lost, 1), 'руб.')
    print('Успешных попыток: ', successes)
    print('Вероятность выигрыша: ', round((successes/trials) * 100, 1), '%')

if __name__ == '__main__':
    main()
```

## Ответы на вопросы

**Выгодна ли игра игроку?**

A = "Игрок получил выигрыш"

H1 = "Выпало обычное число"

H2 = "Выпало число из списка"

P(A) = P(H1)*P(A/H1) + P(H2)*P(A/H2) = 0.795 * 0 + 0.204 * 100 = 20.4%

Данное значение совпадает со значением, который я получила моделированием с помощью метода Монте-Карло.

Делала расчет из следующих соображений: 

Всего чисел, которые могут привести к каким-либо выигрышам, 185. Остальных - 204. Всего 999. То есть 204/999 = 0.204, 795/999 = 0.795

Таким образом, вероятность слишком мала, чтобы рисковать деньгами. Данная игра не является выгодной для игрока. В следующем пункте я оценила среднее значение выигрыша и проигыша, что тоже доказывает невыгодность подобной игры.
