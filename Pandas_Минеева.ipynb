## Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
df_power = pd.read_csv('power.csv')
df_power.head()
```

```python
df_power_1 = df_power[(df_power['year'] >= 2005) & (df_power['year'] <= 2010) & (df_power['quantity'] > 0) & (df_power['category'].isin([4, 12, 21])) & (df_power['country'].isin(['Latvia', 'Lithuania', 'Estonia']))]
df_power_1.sort_values(by='quantity', ascending=False).head()
```

```python
df_power_1.info()
```

```python
sum = df_power_1['quantity'].sum()
print(f'Суммарное потребление стран Прибалтики: {sum}')
```
